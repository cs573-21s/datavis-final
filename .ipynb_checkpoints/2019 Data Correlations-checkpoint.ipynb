```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import lines, rcParams, cycler
import numpy as np, random, operator, pandas as pd
import copy
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
# Read in file
data = pd.read_csv('2019data.csv')
```

```python
grr = pd.plotting.scatter_matrix(data, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```python
print(data.describe())
datanocountries = data.copy()
del datanocountries['Country']
del datanocountries['Rank']
del datanocountries['Score']
print()
print("Correlation Coefficients:")
print(np.corrcoef(np.transpose(datanocountries)))
```
